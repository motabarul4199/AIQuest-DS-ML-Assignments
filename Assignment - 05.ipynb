```python
li = [110,105,115,120,110,130,150,100,105]
```

```python
li
```

```python
import pandas as pd
df = pd.DataFrame(li, columns=['price'])
df
```

Normalization / Min_Max_Scaler

```python
min_price = df.price.min()
min_price
```

```python
max_price = df.price.max()
max_price
```

```python
df['price_norm_manual'] = (df.price - min_price) / (max_price - min_price)
df.head()
```

Sklearn Normalization / Min_Max_Scaler

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
```

```python
df['price_norm_sklearn'] = mms.fit_transform(df[['price']])
```

```python
df.head()
```

Standardization

```python
mean_price = df.price.mean()
std = df.price.std()
```

```python
df['price_std_manual'] = (df.price - mean_price) / std #N std
```

```python
df.head()
```

Sklearn Standardization

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
```

```python
df['price_std_sklearn'] = scaler.fit_transform(df[['price']]) #n-1 best practice
```

```python
df.head()
```

```python
std_skl_n = df.price.std(ddof=0) #ddof means the Delta Degrees of Freedom
std_skl_n
```

```python
df['price_std_manual2'] = (df['price'] - mean_price) / std_skl_n
```

```python
df
```

Log Transformation

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
```

```python
ft = FunctionTransformer(np.log1p)
df['price_ft'] = ft.fit_transform(df[['price']])
```

```python
df
```

just Numpy

```python
df['price_logt_np'] = np.log1p(df[['price']])
df.head()
```

Robust Scaler

```python
from sklearn.preprocessing import RobustScaler
rs = RobustScaler()
```

```python
df['price_rs_sklearn'] = rs.fit_transform(df[['price']])
```

```python
df.head()
```

Max Absolute Scaler

```python
from sklearn.preprocessing import MaxAbsScaler
mas = MaxAbsScaler()
```

```python
df['mas_sklearn'] = mas.fit_transform(df[['price']])
```

```python
df.head()
```
